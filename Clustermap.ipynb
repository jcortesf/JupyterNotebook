# Clustermap
## Importar librerías requeridas

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

## Cargar dataset
El cargado de datos en este caso puede requerir pre-procesamiento. Es necesario contar con un registro de todas las combinaciones de nodos en la red, para verificar si existe una conexión entre ellos o no. Luego, dependiendo del formato inicial de los datos, se deben procesar para recibirse como datos tabulares. en este ejemplo, se lee un archivo CSV donde cada ítem es combinaciones entre nodos, y el campo `"edge"` específica el peso de la conexión. Con el uso de `pivot`se puede indexar el dataset para usarlo en forma de matriz.

```python
adyacencia_2 = pd.read_csv("dataset/graph_2.csv").pivot("source","target","edge")
adyacencia_2
```

## Clustermap
El método clustermap hace un clustering jerárquico de los datos en función de la matriz que uno le entrega. En base al clustering, dibuja el dendograma que vemos arriba y a un lado de la matriz de adyacencia.

```python
sns.clustermap(adyacencia_2, cmap="vlag", center=0, linewidths=.5, figsize=(10,10))
```

## Sintesis

    - El Clustermap es un tipo de visualización que combina las fortalezas de dos tipos de visualizaciones distintas.
    - El Clustermap combina la matriz de adyacencia con un dendograma, lo que permite visualizar la estructura jerárquica de los nodos de la red.
    - Es posible hacer esta visualización con la librería Seaborn.
