# Gráfico de Dispersión y de Burbujas

## A. Importar librerías
A continuación se importan las librerías necesarias para crear las visualizaciones.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

## B. Cargar dataset
El dataset que se utilizará corresponde a propinas dadas en un restaurant

```python
tips = sns.load_dataset("tips")
```

```python
# Cuantas observaciones hay en el dataset?
tips.describe()
```

# C. Gráfico de Dispersión

## C.1 Usando matplotlib

```python
# Se crea un gráfico de dispersión donde el eje X muestra el monto total de la cuenta en el restaurante
# y el eje Y muestra la propina que se dio al mesero(a)

plt.scatter(tips.total_bill, tips.tip)
```

```python
# marker [ '*' | '+' | ',' | '.' | '1' | '2' | '3' | '4' | 's' | 'o' ]
# alpha controla transparencia

plt.scatter(tips.total_bill, tips.tip, color="red", marker="s", alpha=0.3)
plt.title("Monto cuenta vs propina")
plt.xlabel("Monto Cuenta")
plt.ylabel("Propina")
plt.show()
```

## C.2 Gráfico de puntos con seaborn

```python
ax = sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
# Se cambian algunos atributos
sns.set()
sns.set_style("white")

ax = sns.scatterplot(x="total_bill", y="tip", data=tips, color="red", marker="s", alpha=0.3)
```

# D. Gráfico de Burbujas
El gráfico de burbujas es un caso especial del gráfico de dispersión donde además se utiliza los canales de tamaño y color de los puntos presentados.

## D.1 Gráfico de burbujas usando seaborn

```python
# Se transforma alunas columnas de valores categóricas en númericas
smoker_num = tips.smoker.astype("category").cat.codes
time_num = tips.smoker.astype("category").cat.codes
```

```python
# Se pueden modificar varios atributos para hacer bubbleplot en seaborn
# Tamaño de los puntos con **size**
# Color de los puntos con **hue**

sns.set_style("whitegrid")

ax = sns.scatterplot(x="total_bill", y="tip",
                    hue=time_num, size="size",
                    sizes=(20,250), hue_norm=(0,1.5), alpha=0.7, cmap="Set1",
                    legend="full", data=tips)
ax.set_title("Cuenta vs Propina \n (Tamaño de burbuja: gente en mesa) \n (Color: cena vs almuerzo)")
ax.set_xlabel("Monto cuenta")
ax.set_ylabel("Propina")
```
