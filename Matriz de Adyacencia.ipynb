# Matriz de adyacencia
## Importar librerías requeridas
Se Importan los modulos necesarios para generar este tipo de visualización.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Cargar dataset
El cargado de datos en este caso puede requerir pre-procesamiento. Es necesario contar con un registro de todas las combinaciones de nodos en la red, para verificar si existe una conexión entre ellos o no. Luego, dependiendo del formato inicial de los datos, se deben procesar para recibirse como datos tabulares. En este ejemplo, se leen dos archivos CSV donde cada ítem son combinaciones entre nodos, y el campo **"edge"** especifíca si existe una conexión o no. Con el uso de **pivot** se puede indexar el dataset para usarlo en forma de matriz.

```python
# Se carga el dataset mediante pandas
adyacencia_1 = pd.read_csv("dataset/graph_1.csv").pivot("source","target","edge")
adyacencia_1
```

De forma similar, pueden existir casos donde las conexiones tienen datos en sí, como un valor cuantitativo. El siguiente **dataset** muestra este caso, donde ahora hay valores numéricos para cada par de conexiones.

```python
# Se carga el dataset mediante pandas
adyacencia_2 = pd.read_csv("dataset/graph_2.csv").pivot("source","target","edge")
adyacencia_2
```

## Matriz de adyacencia
**Seaborn** no provee una función específica de matriz de adyacencia. Pero, esta puede verse como un caos específico de mapa de calor, por lo que podemos hacer uso de la misma función **sns.heatmap**.

```python
# plt.rcParams['figure.figsize'] permite especificar las dimensiones del gráfico creado
plt.rcParams['figure.figsize'] = (15.0,8.0)
# sns.heatmap recibe el dataset indexado y crea un mapa de calor.
# Se puede específicar paleta con cmap y mantener cuadrado el mapa con square.
ax = sns.heatmap(adyacencia_1, cmap=sns.light_palette("black"), linewidths=.5, square=True)
```

```python
# sns.heatmap recibe el dataset indexado con valores cuantitativos y crea un mapa de calor
ax = sns.heatmap(adyacencia_2, linewidths=.5, square=True)
```
