# Diagrama Nodo-Enlace
## Importar librerías requeridas
Se importan las librerías necesarías para generar las visualizaciones

```python
import matplotlib.pyplot as plt
import networkx as nx
import pandas as pd
```

## Carga de Dataset
El dataset corresponde a los personajes de la pelicula Forest Gum, de los cuales los nodos serán los actores y los enlaces indican las veces en que los actores interactuaron en alguna escena.

```python
G = nx.read_gexf("dataset/forrest_gump.gexf")
```

```python
G = nx.relabel_gexf_graph(G)
```

```python
G.number_of_nodes()
```

```python
G.number_of_edges()
```

```python
list(G.edges)[0:10]
```

## Generar visualización de nodo-enlace

```python
# este corresponde al algoritmo de posicionamiento de nodos
pos = nx.spring_layout(G)
# ahora llamamos el método que dibuja la red
nx.draw(G, pos, node_color="b", node_size=40, with_labels=False)
```

```python
# este corresponde al algoritmo de posicionamiento de nodos
pos = nx.spring_layout(G, seed=2)
# aumentaremos el tamaño de la figura para poder poner etiquetas de nodos sin solapamiento
plt.figure(3,figsize=(10,10))
# ahora llamamos el método que dibuja la red
nx.draw(G, pos, node_color="b", node_size=50, font_size=8, with_labels=True)
```

```python
# Le daremos realce al nodo de los protagonistas: Forrest y Jenny
pos = nx.spring_layout(G, seed=2)
plt.figure(3,figsize=(10,10))

# ahora llamamos el método que dibuja la red
nx.draw(G, pos, node_color="b", node_size=50,font_size=8, with_labels=True)
options = {"node_size": 200, "node_color": "r"} 
nx.draw_networkx_nodes(G, pos, nodelist=['FORREST','JENNY'], **options)
labels = nx.draw_networkx_labels(G.subgraph(['FORREST','JENNY']), pos, font_color='yellow') 
plt.show()
```

## Recomendaciones adicionales
Si bien networkx cumple la funcion de dibujar grafo básicos, su propósito es más bien el análisis de grafos y no el dibujo de grafos.
Existen herramientas especializadas para dibujar grafos como Gephi (https://gephi.org/), Cytoscape (https://cytoscape.org/) o Graphviz (https://www.graphviz.org/), las que recomendamos para dibujar diagramas nodo-enlace de mayor complejida (con mayo número de nodos y enlaces)
