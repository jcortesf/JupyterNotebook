# Gráficos de Puntos y de Líneas

## Importar librerías requeridas
En la siguiente casilla importaremos los módulos necesarios para crear las visualizaciones.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

## Cargar dataset
El dataset corresponde  a observaciones de resonancia magnetica (fmri) hecha a diferentes pacientes (columna subject)

```python
fmri = sns.load_dataset("fmri")
```

```python
# Cuantas observaciones hay en el dataset?
fmri.describe()
```

```python
fmri.head(5)
```

```python
# Para poder graficar una sola serie de puntos
# (una observacion en cada momento en el tiempo) se va a elegir:
# un paciente: s13
# un evento observado: cue (podría ser 'stim')
# una region del cerebro observada: frontal (podría ser 'parietal')

subject13 = fmri.loc[(fmri.subject == 's13') & (fmri.event == 'cue') & (fmri.region == 'frontal')]
print(subject13)
```

```python
# Se ordena las filas en base a la columna timepoint
subject13 = subject13.sort_values(by='timepoint')
print(subject13)
```

# Gráfico de puntos
## 1 Usando matplotlib

```python
# Se grafica la señal en 19 puntos en el tiempo del usuario s13
sns.set()
sns.set_style("whitegrid")

plt.plot(subject13.timepoint, subject13.signal, marker = 'o', linestyle='')
```

```python
# marker [ '*' | '+' | ',' | '.' | '1' | '2' | '3' | '4' | 's' | 'o' ]
plt.plot(subject13.timepoint, subject13.signal
, markersize=10, color='skyblue',marker = '*', linestyle='')
```

## 2 Usando seaborn

```python
# Grafico de los datos subject13 con seaborn
ax = sns.lineplot(x = "timepoint", y = "signal", data = subject13, linewidth = 0, marker = 'o')
```

```python
# Se usará todos los datos asociados a un mismo timepoint ( no sólo subject 13)
# El marker ahora es un promedio de las señales en un mismo timepoint:
# se puede tener barras de desviación estándar asociados
# Esto es muy fácil de hacer con seaborn
# Notar que para el lineplot de seaborn no hay que ordenar los datos previamente
ax = sns.lineplot(x = "timepoint", y = "signal", color = "green",
data = fmri, linewidth = 0, marker = 's', err_style = "bars")
```

```python
# Y con banda de error en lugar de barras de error
ax = sns.lineplot(x="timepoint", y="signal", color="green"
,data=fmri, linewidth=0, marker = 's', err_style="band")
```
